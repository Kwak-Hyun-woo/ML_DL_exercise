```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
```

# 가장 연관된 특징들 추출

```python
iris = datasets.load_iris()
X = iris.data
```

# 표준화하기

```python
from sklearn import preprocessing 

std = preprocessing.StandardScaler()
X_std = std.fit_transform(X)
X_std[:5]
```

```python
data = pd.DataFrame(X_std, columns=iris.feature_names)
```

```python
data
```

# PCA 수행

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
result = pca.fit_transform(data)

result_df = pd.DataFrame(result, columns = ["x", "y"])
```

# 시각화

```python
target = pd.DataFrame(iris.target, columns=['type'])
target['type'] = target['type'].apply(lambda x: iris.target_names[x])
target
merged = pd.concat([result_df, target], axis=1)
```

```python
import matplotlib.pyplot as plt
import matplotlib as mlp
import seaborn as sns

%matplotlib inline

# font 정의
mlp.rcParams['font.size'] = 20
mlp.rcParams['font.family'] = 'Nanum Gothic'

# 시각화
plt.figure(figsize=(16, 9))
sns.set_palette(sns.color_palette("muted"))

sns.scatterplot(merged['x'], 
                     merged['y'], 
                     
                     s=100,
                     palette=sns.color_palette('muted', n_colors=3),
                    )
plt.title('Iris')
plt.show()
```

# KMeans 알고리즘 수행

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3, n_jobs =4, random_state = 1)
km.fit(result)
```

```python
print(km.cluster_centers_)
```

```python
new_labels = km.labels_
# Plot the identified clusters and compare with the answers
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(merged['x'], 
                     merged['y'], 
                     c=iris.target, 
                     cmap='gist_rainbow')

axes[1].scatter(result[:, 0], result[:, 1], c=new_labels, cmap='jet',
edgecolor='k', s=150)

axes[0].set_xlabel('Sepal length', fontsize=18)
axes[0].set_ylabel('Sepal width', fontsize=18)

axes[1].set_xlabel('Sepal length', fontsize=18)
axes[1].set_ylabel('Sepal width', fontsize=18)

axes[0].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[1].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)

axes[0].set_title('Actual', fontsize=18)
axes[1].set_title('Predicted', fontsize=18)
```

# hierarchical_clustering

```python
result
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

linked = linkage(result, 'single')
```

```python
iris.target
```

```python
labelList = iris.target

plt.figure(figsize=(30, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            show_leaf_counts=False,
           leaf_font_size=20,
            leaf_rotation= 360
          )
plt.show()
```
